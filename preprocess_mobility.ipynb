```python
import pandas as pd
import numpy as np
```

```python
mobility = pd.read_csv("./raw data/2020_CA_Region_Mobility_Report.csv")
```

#### need for mobility

```python
mobility = mobility.drop(['country_region_code','country_region','sub_region_2','census_fips_code','iso_3166_2_code','date','place_id'], axis = 1)
```

```python
mobility.rename(columns = {'sub_region_1':'sub_region'}, inplace = True) 
mobility.rename(columns = {'retail_and_recreation_percent_change_from_baseline':'retail_and_recreation'}, inplace = True) 
mobility.rename(columns = {'grocery_and_pharmacy_percent_change_from_baseline':'grocery_and_pharmacy'}, inplace = True) 
mobility.rename(columns = {'parks_percent_change_from_baseline':'parks'}, inplace = True) 
mobility.rename(columns = {'parks_percent_change_from_baseline':'parks'}, inplace = True) 
mobility.rename(columns = {'transit_stations_percent_change_from_baseline':'transit_stations'}, inplace = True) 
mobility.rename(columns = {'workplaces_percent_change_from_baseline':'workplaces'}, inplace = True) 
mobility.rename(columns = {'residential_percent_change_from_baseline':'residential'}, inplace = True) 
```

```python
mobility["metro_area"].fillna("null", inplace = True) 
mobility["retail_and_recreation"].fillna(0, inplace = True)
mobility["grocery_and_pharmacy"].fillna(0, inplace = True) 
mobility["parks"].fillna(0, inplace = True) 
mobility["transit_stations"].fillna(0, inplace = True) 
mobility["workplaces"].fillna(0, inplace = True) 
mobility["residential"].fillna(0, inplace = True) 
mobility = mobility.dropna(axis = 0,how ='any')
```

```python
#check if there is any other null values
mobility.isnull().sum()
```

```python
mobility["province"] = "ON"
```

```python
mobility = mobility.loc[mobility['sub_region'] == 'Ontario']
```
