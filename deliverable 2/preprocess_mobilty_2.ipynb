```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./raw_data/mobility_report.csv")
df
```

```python
def set_mobility(x):
    if x == 'Ottawa Division':
        return 1
    elif x == 'Regional Municipality of Durham':
        return 2
    elif x == 'Regional Municipality of Halton':
        return 3
    elif x == 'Regional Municipality of Peel':
        return 4
    elif x == 'Regional Municipality of York':
        return 5
    else:
        return 6
```

```python
df['mobility_key'] = df['subregion'].apply(set_mobility)
df
```

```python
df.to_csv("mobility.csv", index=False)
```
