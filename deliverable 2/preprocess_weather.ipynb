#### handle duplicaes , null values and rename the columns

```python
df = df.drop_duplicates()
```

```python
df.rename(columns = {'Max Temp (°C)':'daily_high_temperature','Min Temp (°C)':'daily_low_temperature','Total Precip (mm)':'precipitation','Station Name':'location'}, inplace = True) 
```

```python
df
```

```python
df = df[['daily_high_temperature','daily_low_temperature','precipitation','Date/Time']]
df = df.dropna(axis = 0,how ='any')
```

```python
df['Date/Time']= pd.to_datetime(df['Date/Time'])
```

```python
df = df.sort_values(by='Date/Time',ascending=True)
df
```

```python
df = df[(df['Date/Time'] >= '10/16/2020') & (df['Date/Time'] <= '02/16/2021')]
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df = df[['weather_key','daily_high_temperature','daily_low_temperature','precipitation']]
df['weather_key'] = df.index + 1
```

```python
df
```

```python
df.to_csv('weather.csv',index=False)
```
